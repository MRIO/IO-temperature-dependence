```python
from pyneuroml.lems import LEMSSimulation
from pyneuroml.lems import generate_lems_file_for_neuroml
import sys
import os
import pprint; pp = pprint.PrettyPrinter(depth=6)
```

```python
#Create file with helper function

sim_id = 'Vm_CHmyfile'
neuroml_file = "nml_c51_onsite.net.nml"
target='net'
duration=1000
dt = 0.025
lems_file_name = 'LEMS_%s.xml'%sim_id 
target_dir = "LEMSfiles"
if not os.path.isdir(target_dir):
    os.mkdir(target_dir)
    
generate_lems_file_for_neuroml(sim_id, 
                                   neuroml_file, 
                                   target, 
                                   duration, 
                                   dt, 
                                   lems_file_name,
                                   target_dir)
```

```python
#Create file manually

sim_id = 'Vm_CMmyfile'
sim_dur_ms = 700
dt = 0.05
ls = LEMSSimulation(sim_id, sim_dur_ms, dt, 'net')
ls.include_neuroml2_file('nml_c51_onsite.net.nml')
   
disp0 = 'display0'
ls.create_display(disp0,"Voltages", "-90", "50")
   
ls.add_line_to_display(disp0, "v", "pop24/0/C51A/v", "1mV", "#ffffff")
    
of0 = 'Volts_file'
ls.create_output_file(of0, "%s.v.dat"%sim_id)
ls.add_column_to_output_file(of0, 'v', "pop24/0/C51A/v")
    
eof0 = 'Events_file'
ls.create_event_output_file(eof0, "%s.v.spikes"%sim_id, format='ID_TIME')
 
ls.add_selection_to_event_output_file(eof0, '0', "pop24/0/C51A", "spike")
  
ls.set_report_file('report.txt')
    
print("Using information to generate LEMS: ")
pp.pprint(ls.lems_info)
print("\nLEMS: ")
print(ls.to_xml())
    
ls.save_to_file()
assert os.path.isfile('LEMS_%s.xml'%sim_id)
```

```python
def validate_lems2(file_name):

    from lxml import etree
    try:
        from urllib2 import urlopen  # Python 2
    except:
        from urllib.request import urlopen # Python 3
        
    schema_file = urlopen("https://raw.githubusercontent.com/LEMS/LEMS/development/Schemas/LEMS/LEMS_v0.7.3.xsd")
    xmlschema = etree.XMLSchema(etree.parse(schema_file))
    print("Validating {0} against {1}".format(file_name, schema_file.geturl()))
    try:
        xmlschema.assertValid(etree.parse(file_name))
    except etree.DocumentInvalid:
        print("Validation error(s):")
        for error in xmlschema.error_log:
            print("  Line {}: {}".format(error.line, error.message))
```
